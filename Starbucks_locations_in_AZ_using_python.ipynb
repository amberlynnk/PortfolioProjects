```python
##import zipcode boundry map into python##

#plotting zip code boundrys 
az_zip_shp.plot()
```

```python
##count the number of starbucks locations in each zip code using spatial overlay 

starbucks_with_az_zip = gpd.sjoin(coords,az_zip_shp, how='inner', op='intersects')

starbucks_with_az_zip

##group by zipcode

starbucks_by_zipcode = starbucks_with_az_zip.groupby('Zipcode',as_index=False).count()[['Zipcode','address']]

starbucks_by_zipcode


```

```python
##change data to integer 

starbucks_by_zipcode['Zipcode'] = starbucks_by_zipcode['Zipcode'].astype(int)
starbucks_by_zipcode['address'] = starbucks_by_zipcode['address'].astype(int)

```

```python
##merge data to create map 

merged_starbucks_zip = pd.merge(az_zip_shp,starbucks_by_zipcode,left_on='Zipcode', right_on='Zipcode')

merged_starbucks_zip.head()
```

```python
##export zip of starbucks locations as color map 

merged_starbucks_zip.plot(column='address',figsize=(10,10))

plt.savefig('module_5.png')
```
